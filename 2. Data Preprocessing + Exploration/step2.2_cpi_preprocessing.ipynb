```python
# CPI data from US Bureau of Labor Statistics
import pandas as pd
cpi_data = pd.read_csv('cpi.csv')
cpi_data.info()
```

```python
cpi_data.head()
```

```python
#Group CPI data by year and calculate the average CPI for each year
annual_cpi = cpi_data.groupby('Year')['Value'].mean().reset_index()
annual_cpi.columns = ['movie_year_of_release', 'average_cpi']
```

```python
annual_cpi.head()
```

```python
annual_cpi.to_csv('US_BLS_1915-1924_Yearly_Average_CPI.csv', index=False)
```
