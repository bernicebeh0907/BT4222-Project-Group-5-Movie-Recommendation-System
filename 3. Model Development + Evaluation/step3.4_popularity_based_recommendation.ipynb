# Find favourite genre of each reviewer

Group all the reviews of each reviewer together. Calculate the favourite genre of each reviewer based on the movie genre that they have rated the highest, amongst all the reviews they provided.

```python
# Check the movie genre format in `movie_genre` column to see if it is a string representation of a list.
# If so, we can parse it.
dg['movie_genre'] = dg['movie_genre'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# Explode the genres to have one genre per row for each review
dg_exploded = dg.explode('movie_genre')

# Group by reviewer and genre, calculating the average rating for each genre
average_genre_ratings = dg_exploded.groupby(['reviewer_name', 'movie_genre'])['reviewer_rating'].mean().reset_index()

# For each reviewer, find their favorite genre (highest average rating)
favorite_genres = average_genre_ratings.loc[average_genre_ratings.groupby('reviewer_name')['reviewer_rating'].idxmax()]

# Display the favorite genre of each reviewer
print(favorite_genres.head())
```

# Calculating the weighted score


We are taking into account 3 criteria to calculate a weighted score for each movie. They are:

*   **movie popularity score**

Strong indicator of audience reception, but it could also be biased towards movies that are marketed well or have a large following, rather than movies that are necessarily critically acclaimed or innovative.

*   **oscar awards won**

Awards won can be a strong indicator of quality, critical acclaim, and artistic value.

*   **revenue per minute**

Good indicator of **financial success relative to runtime** and determines whether the movie was a blockbuster success. It can indicate that the movie's length is well-suited to its audience and **highly demanded amongst audience** .Without this metric, comparing two movies of vastly different lengths could be misleading. A long movie might earn more revenue, but it may not necessarily be more profitable per minute compared to a shorter movie. This metric normalizes the revenue based on the movie's duration, allowing for a fairer comparison.


---

As such we have recommended movies based on popularity by taking into account broad audience appeal, financial success and critical recognition with artistic merit. These 3 metrics are given equal weightage in the calculation of the weighted score to so that movies are recommended in a more holistic manner.

We will recommend each reviewer, **the top 5 movies of their favourite genre, based on the movies' weighted score**.


```python
import re

# Step 1: Clean the movie_title in dg by removing the year and converting to title case
dg['cleaned_movie_title'] = dg['movie_title'].apply(lambda x: re.sub(r' \(\d{4}\)', '', x).title())

# Step 2: Merge dg and df on cleaned movie titles to bring in the total_awards_won column
merged_df = pd.merge(
    dg,
    df[['film_title', 'total_awards_won']],
    left_on='cleaned_movie_title',
    right_on='film_title',
    how='left'
)

# Step 3: Drop the 'film_title' column from the merged DataFrame as it's redundant
merged_df = merged_df.drop(columns=['film_title'])

# Step 4: Fill NaN values in total_awards_won with 0 (no awards won if missing)
merged_df['total_awards_won'] = merged_df['total_awards_won'].fillna(0)

# Step 5: Normalize revenue_per_minute and total_awards_won to the [0, 1] range using Min-Max scaling
min_revenue = merged_df['revenue_per_minute'].min()
max_revenue = merged_df['revenue_per_minute'].max()
merged_df['normalized_revenue_per_minute'] = (merged_df['revenue_per_minute'] - min_revenue) / (max_revenue - min_revenue)

min_awards = merged_df['total_awards_won'].min()
max_awards = merged_df['total_awards_won'].max()
merged_df['normalized_total_awards_won'] = (merged_df['total_awards_won'] - min_awards) / (max_awards - min_awards)

# Step 6: Calculate the weighted score using the normalized values
merged_df['weighted_score'] = (
    merged_df['normalized_revenue_per_minute'] * 0.33 +
    merged_df['movie_popularity'] * 0.33 +
    merged_df['normalized_total_awards_won'] * 0.34
)

# Step 7: Update the recommendation function to return top 5 recommended movies and the genre
def recommend_movies_for_reviewer(reviewer_name):
    # Check if the reviewer exists in favorite_genres
    if reviewer_name not in favorite_genres['reviewer_name'].values:
        return f"Reviewer '{reviewer_name}' not found in the dataset", None

    # Get the reviewer's favorite genre
    favorite_genre = favorite_genres.loc[favorite_genres['reviewer_name'] == reviewer_name, 'movie_genre'].values[0]

    # Filter movies in the merged_df for the identified favorite genre
    movies_in_favorite_genre = merged_df[merged_df['movie_genre'].apply(lambda x: favorite_genre in x)]

    # If no movies are found in the favorite genre, return a message
    if movies_in_favorite_genre.empty:
        return f"No movies found in the '{favorite_genre}' genre for recommendation", None

    # Remove duplicates based on 'cleaned_movie_title'
    movies_in_favorite_genre = movies_in_favorite_genre.drop_duplicates(subset='cleaned_movie_title')

    # Rank the movies in the favorite genre by normalized weighted score
    ranked_movies = movies_in_favorite_genre.sort_values(by='weighted_score', ascending=False)

    # Get the top 5 movies
    top_5_movies = ranked_movies[['cleaned_movie_title', 'weighted_score']].head(5)

    # Format the recommendations as a list of tuples (movie_title, weighted_score)
    recommendations = top_5_movies.apply(
        lambda row: (row['cleaned_movie_title'], f"{row['weighted_score']:.5f}"), axis=1
    ).tolist()

    return favorite_genre, recommendations

```

```python
# Example: Recommend top 5 movies for a specific reviewer. Change the reviewer name to any existing user.
# Here are more example users: Dr_Kruger, jazzest, sol1218, jtindahouse
reviewer_name = 'jtindahouse'  # Replace with actual reviewer name
favorite_genre, recommended_movies = recommend_movies_for_reviewer(reviewer_name)

if isinstance(recommended_movies, str):  # Error message returned
    print(recommended_movies)
else:
    print(f"Top 5 Recommended Movies for {reviewer_name} in the '{favorite_genre}' genre:")
    for i, (movie_title, movie_score) in enumerate(recommended_movies, start=1):
        print(f"{i}. {movie_title} with a score of {movie_score}")
```

# Normalise the revenue per minute and oscars

We have normalised the revenue per minute and oscar awards won in this file. The movie popularity score has already been normalised in preprocessing.  

We **avoided skewed influence by normalizing each metric independently**. We ensured that each component contributes proportionately to the final score. This approach reduces the likelihood of any single metric (especially revenue per minute that is in the larger natural range) to dominate the final weighted score simply because of scale differences.
